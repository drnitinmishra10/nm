### Merge in **Pandas**

```
left = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'], 
                    'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3']}) 
```

```
right = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3','K4'], 
                      'C': ['C0', 'C1', 'C2', 'C3','c4'], 
                      'D': ['D0', 'D1', 'D2', 'D3','c5']})
```

```
left
```

```
right
```

Inner join

```
pd.merge(left, right, how ='inner', on ='Key') 
```

Outer Join

```
pd.merge(left, right, how ='outer', on ='Key') 
```

## Join In Pandas

```
left1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3']}, 
                    index = ['K0', 'K1', 'K2', 'K3']) 
  
right1 = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'], 
                      'D': ['D0', 'D1', 'D2', 'D3']}, 
                      index = ['K0', 'K1', 'K2', 'K3']) 
                        
```

```
left1
```

```
right1
```

```
left1.join(right1)
```
